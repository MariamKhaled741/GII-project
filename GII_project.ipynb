```python
import pandas as pd

df = pd.read_excel("/content/HDR25_Statistical_Annex_GII_Table.xlsx", header=5)

df = df.dropna(axis=1, how='all')

df = df[df['Country'].notna()]

df.columns = df.columns.str.strip()

df.replace(['..', 'b', 'â€“'], pd.NA, inplace=True)

numeric_cols = df.columns.drop('Country')
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

df.head(10)
```

```python
df.columns = [
    "HDI_rank", "Country", "GII_Value", "GII_Rank",
    "Maternal_mortality_ratio", "Adolescent_birth_rate",
    "Share_of_seats_in_parliament", "drop1",
    "Education_Female", "drop2", "Education_Male", "drop3",
    "Labour_Female", "Labour_Male"
]

df = df.drop(columns=["drop1", "drop2", "drop3"])

df = df[df["Country"].notna()]
df = df[df["Country"].str.contains("development") == False]

cols_to_num = df.columns.drop(["Country"])
df[cols_to_num] = df[cols_to_num].apply(pd.to_numeric, errors="coerce")

df.head()
```

```python
for col in cols_to_fill:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)


print(df.isna().sum())
```

```python
df['Education_gap'] = df['Education_Male'] - df['Education_Female']
df['Labour_gap'] = df['Labour_Male'] - df['Labour_Female']
```

```python
df['Education_GPI'] = df['Education_Female'] / df['Education_Male']
df['Labour_GPI'] = df['Labour_Female'] / df['Labour_Male']
```

```python
def hdi_group(rank):
    if rank <= 50:
        return 'Very High'
    elif rank <= 100:
        return 'High'
    elif rank <= 150:
        return 'Medium'
    else:
        return 'Low'

df['HDI_group'] = df['HDI_rank'].apply(hdi_group)
```

```python
df = df[['Country', 'HDI_rank', 'HDI_group', 'GII_Value',
         'Maternal_mortality_ratio', 'Adolescent_birth_rate',
         'Share_of_seats_in_parliament', 'Education_Female', 'Education_Male',
         'Education_gap', 'Education_GPI', 'Labour_Female', 'Labour_Male',
         'Labour_gap', 'Labour_GPI']]
```
